# DATE TIME

```python
data["Order Date"] = pd.to_datetime(data["Order Date"], format="%d-%m-%Y")
data["Ship Date"] = pd.to_datetime(data["Ship Date"], format="%d-%m-%Y")
```

```python
import holidays
# Lấy danh sách ngày lễ US
us_holidays = holidays.US(years=[2011,2012,2013,2014])

data['IsHoliday'] = data['Order Date'].dt.date.isin(us_holidays).astype(int)

# Nếu chưa có IsWeekend
data['IsWeekend'] = data['Order Date'].dt.dayofweek >= 5  
data['IsWeekend'] = data['IsWeekend'].astype(int)

# Tạo HolidayScore
data['HolidayScore'] = data['IsHoliday'] + data['IsWeekend']
```

```python
data[["IsHoliday", "IsWeekend", "HolidayScore"]].value_counts()
```

```python
# Tạo cột DaysToShip
data['DaysToShip'] = (data['Ship Date'] - data['Order Date']).dt.days
```

# FINISH

```python
data = data.rename(
    columns={
       "Row ID": "RowID",
       "Order ID": "OrderID",
       "Product ID": "ProductID",
       "Order Date": "OrderDate",
       "Ship Date": "ShipDate",
       "Ship Mode": "ShipMode",
       "Customer ID": "CustomerID",
       "Customer Name": "CustomerName",
       "Postal Code": "PostalCode",
       "Sub-Category": "SubCategory",
       "Product Name": "ProductName"
    }
)
```

```python
data.info()
```

```python
data = data.drop(columns=["Country"])
```

```python
for col in data.select_dtypes(include='object'):
    max_len = data[col].str.len().max()
    print(f"Column '{col}' max length: {max_len}")
```
